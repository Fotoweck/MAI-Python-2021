```python
import math
x=[i*0.1 for i in range(100)]
y=[math.sin(x) for x in x]
y2=[math.cos(x) for x in x]
plt.plot(x,y,'r+',)
plt.plot(x,y2,'g',)
plt.title('График') # подпись
plt.legend()
```

```python
import math
x=[i*0.1 for i in range(100)]
y=[math.sin(x) for x in x]
y2=[math.cos(x) for x in x]
plt.plot(x,y,'r+',label='1' ) # label - подпись к линиям 
plt.plot(x,y2,'g',label='2')
plt.title('График') # подпись
plt.legend()
```

```python
fig,ax=plt.subplots() # создаёт пустую фигуру
ax.plot(x,y)
fig.dpi=100 #величивает изображение
```

```python
fig,ax=plt.subplots() 
fig.set_size_inches(5,5)# размерность в  дюймах
ax.plot(x,y)
```
